```python
import pymupdf4llm
import pathlib
import os
```

```python
def convert_pdf_folder_to_md(input_folder, output_folder):
    # 1. Setup Path
    input_path = pathlib.Path(input_folder)
    output_path = pathlib.Path(output_folder)

    # Buat folder output jika belum ada
    output_path.mkdir(parents=True, exist_ok=True)

    # 2. Ambil semua file PDF
    pdf_files = list(input_path.glob("*.pdf"))

    if not pdf_files:
        print(f"Tidak ada file PDF di folder: {input_folder}")
        return

    print(f"Mulai konversi {len(pdf_files)} file...")

    # 3. Loop Konversi
    for pdf_file in pdf_files:
        try:
            print(f"Processing: {pdf_file.name}...", end=" ")

            # Convert PDF ke Markdown
            md_text = pymupdf4llm.to_markdown(str(pdf_file))

            # Simpan file .md
            output_filename = pdf_file.stem + ".md"
            save_path = output_path / output_filename

            with open(save_path, "w", encoding="utf-8") as f:
                f.write(md_text)

            print("✅ Selesai")

        except Exception as e:
            print(f"❌ Gagal: {e}")
```

```python
if __name__ == "__main__":
    # Ubah path folder di sini
    FOLDER_INPUT = "../knowledge-base/raw"   # Folder asal PDF
    FOLDER_OUTPUT = "../knowledge-base/processed-pymupdf"  # Folder tujuan Markdown
    
    convert_pdf_folder_to_md(FOLDER_INPUT, FOLDER_OUTPUT)
```
