```python
import os
import glob
import re
```

```python
text_to_remove = [
    "<!-- image -->",
    "<!-- formula-not-decoded -->",
]
```

```python
FOLDER_PATH = "knowledge-base/processed"
OUTPUT_PATH = "knowledge-base/processed-1"
os.makedirs(OUTPUT_PATH, exist_ok=True)

list_file = glob.glob(os.path.join(FOLDER_PATH, "*.md"))

if not list_file:
    print(f"Tidak ada file .md di folder {FOLDER_PATH}")
else:
    print(f"Ditemukan {len(list_file)} file. Memulai pembersihan...\n")

    for file_path in list_file:
        filename = os.path.basename(file_path)
        new_file_path = os.path.join(OUTPUT_PATH, filename)

        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()

        # ✅ PERBAIKAN: Gunakan variabel yang sama agar perubahannya bertumpuk
        for text in text_to_remove:
            content = content.replace(text, "")

        # Rapikan spasi setelah semua teks pengganggu dihapus
        cleaned_content = "\n".join([line.rstrip() for line in content.splitlines()])
        cleaned_content = cleaned_content.strip()

        with open(new_file_path, "w", encoding="utf-8") as f:
            f.write(cleaned_content)

        print(f"Bersih: {filename}")

    print("\nSelesai! Semua file Anda sekarang sudah bersih.")
```
